#### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
cv_scores = []
for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(
        knn,
        X_train_scaled,
        y_train,
        cv=5,
        scoring="recall"
    )
    cv_scores.append(scores.mean())
    print("K:", k, "CV Recall:", round(scores.mean(), 4))
best_k = np.argmax(cv_scores) + 1
print("Best K:", best_k)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import *
KNN=KNeighborsClassifier(n_neighbors=best_k)
KNN.fit(X_train_scaled,y_train)
y_kpred=KNN.predict(X_test_scaled)
```

#### Task 3: Suggestions to the Hospital to Prevent Life-Threatening Heart Disease Outcomes

##### 1) Prioritize Early Screening for High-Risk Indicators
 Based on the correlation heatmap and feature analysis, chest_pain_type, tha (stress test result), `oldpeak_eq_st_depression` (ST depression), and num_major_vessels showed the strongest association with heart disease presence. The hospital should ensure these tests are 
part of routine cardiac screening, especially for patients presenting with any chest discomfort, 
rather than reserving them only for high-suspicion cases.

##### 2) Use the Model as a Triage/Support Tool, Not a Replacement for Doctors
Gradient Boosting (baseline) achieved 100% recall on the test set, meaning it did not miss any actual heart disease cases in evaluation. 
This makes it well-suited as a triage support 
tool — flagging high-risk patients for priority review by a cardiologist — rather than a 
stand-alone diagnostic tool. Final diagnosis should always remain with a qualified physician.

##### 3) Monitor Exercise-Induced Angina and Max Heart Rate Closely
Patients showing exercise-induced angina or unusually low maximum heart rate achieved during stress tests showed 
a visibly higher rate of heart disease in the EDA. 
The hospital should flag these patients for closer follow-up, even if other vitals appear normal.

##### 4) Track ST Depression (oldpeak) as a Key Risk Signal
Boxplots showed a clear separation between oldpeak values for patients with and without heart disease. 
This is a low-cost, easily measurable indicator during a standard ECG and should be weighted heavily in initial risk assessment, especially for patients who cannot immediately access more expensive imaging tests.

##### 5) Build Age and Sex-Specific Awareness Programs
 The data showed heart disease risk varying notably with age and sex. The hospital could run targeted screening camps or awareness drives for older patients and higher-risk demographic groups identified from local patient data, rather than a one-size-fits-all outreach approach.

##### 6) Recommend Periodic Model Retraining
 As more patient data is collected, the hospital's data team should periodically retrain the model. This project used a limited dataset (180 records); model reliability will improve significantly as more patient records become available, and predictions should be 
revisited/recalibrated at least annually.

##### 7) Combine Model Alerts with Human Clinical Judgment
 False positives (healthy patients flagged as at-risk) should lead to a confirmatory low-cost test rather than immediate invasive procedures, given the model's precision (~84%) leaves some room for false alarms. This balances patient safety (via high recall) with avoiding 
unnecessary anxiety/cost for healthy patients.

#### Model Comparison Report

```python
baseline_models = ["Random Forest", "KNN (baseline)", "SVM", "Naive Bayes",
                    "Gradient Boosting", "Logistic Regression", "Decision Tree"]

baseline_comparison = pd.DataFrame([
    {"Model": m, **{k: (f"{v:.2f}%" if k != "AUC" else round(v, 4)) for k, v in results_store[m].items()}}
    for m in baseline_models
]).sort_values("AUC", ascending=False, key=lambda s: s.astype(str).str.rstrip('%').astype(float) if s.name != "AUC" else s)

print("Baseline Model Performance (Default Parameters)")
print(baseline_comparison.to_string(index=False))
```

```python
tuned_pairs = [
    ("Gradient Boosting", "Gradient Boosting (Tuned)"),
    ("KNN (baseline)", "KNN (Tuned)"),
    ("Random Forest", "Random Forest (Tuned)"),
]
rows = []
for base_name, tuned_name in tuned_pairs:
    base_recall = results_store[base_name]["Recall"]
    tuned_recall = results_store[tuned_name]["Recall"]
    result = "Improved" if tuned_recall > base_recall else ("Unchanged" if tuned_recall == base_recall else "Worsened - baseline kept")
    rows.append({
        "Model": base_name.replace(" (baseline)", ""),
        "Baseline Recall": f"{base_recall:.2f}%",
        "Tuned Recall": f"{tuned_recall:.2f}%",
        "Result": result
    })
tuning_comparison = pd.DataFrame(rows)
print("Hyperparameter Tuning Results (GridSearchCV, scoring='recall')")
print(tuning_comparison.to_string(index=False))
```

recall improvement from GridSearchCV (87.5% → 93.75%). Gradient Boosting's recall dropped 
sharply after tuning (100% → 81.25%), along with its accuracy, precision, and F1. Random 
Forest's recall stayed unchanged (93.75%) though its other metrics improved slightly. Since 
recall was the priority metric, only the KNN and Random Forest tuned results were worth 
considering as alternatives to baseline — Gradient Boosting's baseline remained the strongest 
choice overall, likely due to cross-validation instability on a training set of only 144 rows.

##### Challenge 3 — Choosing the Right Metric
Accuracy alone is misleading in healthcare, since missing a real case is worse than a false alarm. Prioritized Recall (scoring="recall") throughout model selection and tuning.

##### Challenge 4 — Categorical Encoding
thal was the only text-based categorical column. One-hot encoded it (drop_first=True) since it's a nominal category with no natural order — normal, fixed_defect, and reversible_defect aren't ranked relative to each other, so LabelEncoder's 0/1/2 mapping would have wrongly implied an order that could bias distance- and coefficient-based models (Logistic Regression, KNN, SVM).

##### Challenge 5 — Feature Scaling
Logistic Regression, KNN, and SVM needed scaled data; tree-based models (Decision Tree, Random Forest, Gradient Boosting) and Naive Bayes didn't. Applied StandardScaler (fit on train only, transformed on test) and used it selectively per model.

##### Challenge 6 — SVM Scaling Bug
 AUC calculation initially used unscaled X_test while the model was trained on scaled data, giving a meaningless AUC of 0.50. Fixed by using X_test_scaled consistently, restoring AUC to 0.9469.

##### Best Model for Production
 Gradient Boosting (Baseline, default parameters) — highest recall (100%) and AUC (0.9531) of any model, baseline or tuned. Notably the only case where the untouched baseline clearly outperformed its own tuned version, since GridSearchCV optimized for recall on the 
training folds but that gain didn't transfer to the held-out test set.

##### Limitations & Future Work
This model was trained on only 180 patient records, which limits how confidently these 
results generalize — a larger, multi-site dataset would meaningfully improve reliability. 

Future improvements could include: replacing the single train/test split with full 
cross-validation-based model selection for a more stable comparison across models; expanding 
GridSearchCV to cover all 7 models rather than just the top 3; and exploring ensemble/stacking 
approaches (e.g., combining Gradient Boosting and Random Forest predictions) to reduce 
variance further.
