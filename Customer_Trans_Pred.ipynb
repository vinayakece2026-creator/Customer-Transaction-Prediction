#### Task 2 : Create a predictive model which will help the bank to identify which customer will make transactions in future

```python
X=data.drop(["ID_code","target"],axis=1)
```

```python
X.head()
```

```python
y=data["target"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sd=StandardScaler()
X_train=sd.fit_transform(X_train)
X_test=sd.transform(X_test)
```

#### Logistic Regression

```python
import warnings
warnings.filterwarnings('ignore') 
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(class_weight="balanced",max_iter=1000)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,f1_score,recall_score,roc_auc_score
print("Confusion matrix :",confusion_matrix(y_test,y_pred))
print("Accuracy :",accuracy_score(y_test,y_pred))
print("precision score :",precision_score(y_test,y_pred))
print("recall score :",recall_score(y_test,y_pred))
print("f1_score :",f1_score(y_test,y_pred))
```

```python
y_prob = model.predict_proba(X_test)[:,1]
auc = roc_auc_score(y_test, y_prob)
print("AUC :", auc)
```

### KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
for k in range(1,21):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    pred=knn.predict(X_test)
    acc_score=accuracy_score(y_test,pred)
    print("K :",k ,"ACC :",acc_score)
```

```python
from sklearn.neighbors import KNeighborsClassifier
Knn=KNeighborsClassifier(n_neighbors=7)
Knn.fit(X_train,y_train)
y_kpred=Knn.predict(X_test)
```

```python
from sklearn.metrics import *
print("Confusion matrix",confusion_matrix(y_test,y_kpred))
print("Accuracy :",accuracy_score(y_test,y_kpred))
print("Precision score :",precision_score(y_test,y_kpred))
print("recall score :",recall_score(y_test,y_kpred))
print("F1_score :",f1_score(y_test,y_kpred))
```

```python
from sklearn.metrics import roc_auc_score
y_prob1 = Knn.predict_proba(X_test)[:,1]
print("AUC :", roc_auc_score(y_test, y_prob1))
```

### Decsion Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(max_depth=10,class_weight='balanced',
    random_state=42)
dtree.fit(X_train,y_train)
y_dpred=dtree.predict(X_test)
```

```python
from sklearn.metrics import *
print("Confusion Matrix:",confusion_matrix(y_test, y_dpred))
print("Accuracy :", accuracy_score(y_test, y_dpred))
print("Precision :", precision_score(y_test, y_dpred))
print("Recall :", recall_score(y_test, y_dpred))
print("F1 Score :", f1_score(y_test, y_dpred))
y_prob2 = dtree.predict_proba(X_test)[:,1]
print("AUC :", roc_auc_score(y_test, y_prob2))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rmodel = RandomForestClassifier(class_weight='balanced',
    n_estimators=20,
    random_state=42,
    n_jobs=-1
)
rmodel.fit(X_train, y_train)
y_rpred = rmodel.predict(X_test)
```

```python
from sklearn.metrics import *
print("Confusion Matrix:",confusion_matrix(y_test, y_rpred))
print("Accuracy :", accuracy_score(y_test, y_rpred))
print("Precision :", precision_score(y_test, y_rpred))
print("Recall :", recall_score(y_test, y_rpred))
print("F1 Score :", f1_score(y_test, y_rpred))
y_prob3 = rmodel.predict_proba(X_test)[:,1]
print("AUC :", roc_auc_score(y_test, y_prob3))
```

#### XG Boost

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=6,
    scale_pos_weight=9,
    random_state=42
)
xgb.fit(X_train, y_train)
pred_xgb = xgb.predict(X_test)
```

```python
from sklearn.metrics import *
print("Confusion Matrix:\n", confusion_matrix(y_test,pred_xgb))
print("Accuracy :", accuracy_score(y_test,pred_xgb))
print("Precision :", precision_score(y_test,pred_xgb))
print("Recall :", recall_score(y_test,pred_xgb))
print("F1 Score :", f1_score(y_test,pred_xgb))
y_prob5 = xgb.predict_proba(X_test)[:,1]
print("AUC :", roc_auc_score(y_test,y_prob5))
```

#### Challenges Faced and Techniques Used

#### Challenge 1 — Class Imbalance
Problem  : 90% class 0, 10% class 1. KNN=0 caught, RF=2 caught.
Technique: class_weight='balanced' for LR, DT, RF.
            scale_pos_weight=9 for XGBoost.
#### Challenge 2 — Accuracy is Misleading
Problem  : KNN/RF showed 89% accuracy but F1=0%. Predicted all as class 0.
Technique: Used AUC, Recall, F1 as primary metrics.
#### Challenge 3 — Anonymous Features
Problem  : 200 features named var_0 to var_199. No domain knowledge.
Technique: StandardScaler applied equally to all 200 features.
#### Challenge 4 — Large Dataset
Problem  : 200,000 rows × 200 features. Heavy computation.
Technique: n_jobs=-1 for parallel processing.
#### Best Model for Production — XGBoost
F1    : 45.44%  (highest — best overall balance)
AUC   : 85.11%  (strong discriminating ability)
Recall: 64.95%  (catches 2,611 out of 4,020 transacting customers)
Reason: Best F1 Score + industry standard for banking tabular data.
